```python
# 2048 GAME

import random


def start_game():
    mat = []
    for i in range(4):
        mat.append([0]*4)
    return mat

# adding the value 2 into the game bord


def add_number_2(mat):

    r = random.randint(0, 3)
    c = random.randint(0, 3)
    while mat[r][c] != 0:
        r = random.randint(0, 3)
        c = random.randint(0, 3)
    mat[r][c] = 2

# current state of the game


def get_current_state(mat):

    # checking for the 2048 into the game
    for i in range(4):
        for j in range(4):
            if mat[i][j] == 2048:
                return "Won"

    # checking for the zero into the game
    for i in range(4):
        for j in range(4):
            if mat[i][j] == 0:
                return "Game Not Over"

    # checking for the row and coloum exect the last row
    for i in range(3):
        for j in range(3):
            if mat[i][j] == mat[i+1][j] or mat[i][j] == mat[i][j]:
                return "Game Not Over"

    # for the last row
    for j in range(3):
        if mat[3][j] == mat[3][j+1]:
            return "Game Not Over"

    # for the last coloumn
    for i in range(3):
        if mat[i][3] == mat[i+1][3]:
            return "Game Not Over"


# COMPRESS Function

def compress(mat):

    new_mat = []
    for i in range(4):
        new_mat.append([0]*4)
    changed = False 
    for i in range(4):
        pos = 0
        for j in range(4):
            if mat[i][j] != 0:
                new_mat[i][pos] = mat[i][j]
                if j != pos :
                    changed = True 
                pos += 1
    return new_mat,changed


# MERGING THE NUMBERS
def merge(mat):
    changed = False
    for i in range(4):
        for j in range(3):
            if mat[i][j] == mat[i][j+1] and mat[i][j] != 0:
                mat[i][j] = mat[i][j]*2
                changed = True 
                mat[i][j+1] = 0
    return mat, changed

### Reverse 
def reverse(mat):

    new_mat = []
    for i in range(4):
        new_mat.append([])
        for j in range(4):
            new_mat[i].append(mat[i][4-j-1])
    return new_mat

# Transpose 
def transpose(mat):

    new_mat = []
    for i in range(4):
        new_mat.append([])
        for j in range(4):
            new_mat[i].append(mat[j][i])

    return new_mat 


# Left Move 
def move_up(grid):
    transposed_grid = transpose(grid)
    new_grid, changed = move_left(transposed_grid)
    final_grid = transpose(new_grid)
    return final_grid, changed

def move_down(grid):
    transposed_grid = transpose(grid)
    new_reverse,changed = move_right(transposed_grid)
    final_grid = transpose(new_reverse)
    return final_grid,changed

def move_right(grid):
    reverse_grid = reverse(grid)
    new_grid,changed = move_left(reverse_grid)
    final_grid = reverse(new_grid)
    return final_grid,changed

def move_left(grid):
    new_grid,changed1 = compress(grid)
    new_grid,changed2 = merge(new_grid)
    changed = changed1 or changed2
    new_grid,change1 = compress(new_grid)
    return new_grid, changed 

mat = start_game()
add_number_2(mat)
print(mat)
val = move_up(mat)
print(val)
```

```python
import math 

def sumOfNatralNumber(num):
    i = 1 
    falg = False 
    while i*(i+1)<num*2:
        x = i*(i+1)
        l = num*2-x 
        k = int(math.sqrt(l))
        if k*(k+1)==l:
            falg = True 
            break
        i+=1
    if falg :
        print("YES")
    else:
        print("NO")
n = int(input())
sumOfNatralNumber(n)
```

```python
def powerset(seq):
    if len(seq) <= 0:
        yield []
    else:
        for item in powerset(seq[1:]):
            yield [seq[0]]+item
            yield item
n,m = map(int,input().split())
avoid = []
for i in range(m):
    a,b = map(str,input().split())
    avoid.append(('$'+a+'$','$'+b+'$'))
lis = [i for i in range(1,n+1)]
ways = []
for way in list(powerset(lis)):
    way_str = '$'+'$'.join([str(j) for j in way])+'$'
    ways.append(way_str)
possible = {}
for i in ways:
    possible[i] = 1
for i in avoid:
    for j in ways:
        if i[0] in j and i[1] in j:
            possible[j] = 0
print(sum(possible.values()))
```
