#### insertion  recursively

```python
class Node :
    def __init__(self, data):
        self.data = data 
        self.next = None
    
### inputs
def takeInput():
    inputList = [int(i) for i in input().split()]
    head = None 
    tail = None 
    for current in inputList:
        if current == -1 :
            break
        newNode = Node(current)
        if head is None:
            head = newNode 
            tail = newNode
        else :
            tail.next = newNode
            tail = newNode
    return head 

### printing the linked List

def printLL(head) :
    while head :
        print(head.data,'-> ',end = '')
        head = head.next 
    print("None")


## length of the linked list 
def length(head):
    count = 0 
    while head :
        count+=1
        head = head.next 
    return count 

### insertion in the linked list 

def insertAt(head,i,data):
    if i < 0 :
        return head
    if i == 0 :
        newNode = Node(data)
        newNode.next = head
        return newNode
    if head is None :
        return None 
    smallHead = insertAt(head.next,i-1,data)
    head.next = smallHead
    return head 



### deletion of the node recursively 
def deletion(head,i):
    if i < 0 :
        return head 
    if head is None :
        return None
    if i == 0 :
        connection = head.next 
        return connection
    smallHead = deletion(head.next,i-1)
    head.next = smallHead
    return head


head = takeInput()
printLL(head)
print(length(head))

head = insertAt(head,5,90)
printLL(head)
head = deletion(head,3)
printLL(head) 

```

### check palindrom linked list 

```python

import sys
sys.setrecursionlimit(10**6)

class Node :
    def __init__(self,data):
        self.data = data 
        self.next = None 

def length(head):
    c = 0 
    while head :
        c+=1
        head = head.next
    return c


def reverse(head):
    prev = None
    current = head
    while current is not None: 
        next = current.next
        current.next = prev
        prev = current
        current = next
    head = prev
    return head 
  

def takeInput():
    inputList = [int(i) for i in input().split()]
    head = None 
    tail = None 
    for i in inputList :
        if i == -1 :
            break

        newNode = Node(i)
        if head is None :
            head = newNode
            tail = newNode
        else :
            tail.next = newNode
            tail = newNode
    return head

def palindrom(head,l) :
    m = l//2
    curr = head 
    Next = curr.next
    if l%2 !=0 :
        c = 0 
        while c < m :
            curr = curr.next 
            Next = curr.next 
            c+=1
        curr.next = None
        tempHead = Next
        rev = reverse(tempHead)

        tempHead = rev 
        curr = head 
        while tempHead :
            if tempHead.data == curr.data :
                tempHead = tempHead.next 
                curr = curr.next 
            else :
                return False
        return True
    else :
        c = 1 
        while c < m :
            curr = curr.next 
            Next = curr.next 
            c+=1
        curr.next = None
        tempHead = Next
        rev = reverse(tempHead)

        tempHead = rev 
        curr = head 
        while tempHead :
            if tempHead.data == curr.data :
                tempHead = tempHead.next 
                curr = curr.next 
            else :
                return False
        return True



def printLL(head) :
    while head :
        print(head.data,end = ' ')
        head = head.next 
    print()

    

t = int(input())
while t > 0 :
    head = takeInput()
    if head :
        l = length(head)
        value = palindrom(head,l)
        printLL(head)
        if value :
            print("true")
        else :
            print(-1)
    else :
        print()
    t-=1 
```
