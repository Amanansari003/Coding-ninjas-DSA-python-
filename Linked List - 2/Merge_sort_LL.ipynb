```python
                    Code : Merge Sort

        
 Given a singly linked list of integers, sort it using 'Merge Sort.'
Note :
No need to print the list, it has already been taken care. Only return the new head to the list.
Input format :
The first line contains an Integer 't' which denotes the number of test cases or queries to be run. Then the test cases follow.

The first and the only line of each test case or query contains the elements of the singly linked list separated by a single space.
Remember/Consider :
While specifying the list elements for input, -1 indicates the end of the singly linked list and hence, would never be a list element
Output format :
For each test case/query, print the elements of the sorted singly linked list.

Output for every test case will be printed in a seperate line.
Constraints :
1 <= t <= 10^2
0 <= M <= 10^5
Where M is the size of the singly linked list.

Time Limit: 1sec
Sample Input 1 :
1
10 9 8 7 6 5 4 3 -1
Sample Output 1 :
 3 4 5 6 7 8 9 10 
```

```python

class Node :
    def __init__(self,data):
        self.data = data 
        self.next = None 
        
def midValue(head):
    slow = head 
    fast = head 
    while fast. next and fast.next.next :
        slow = slow.next 
        fast = fast.next.next  
    return slow



def merge(head):
    if head is None or head.next is None :
        return head 
    mid = midValue(head)
    m = mid.next 
    mid.next = None 
    l = head 
    newHead1 = merge(l)
    newHead2 = merge (m)      
    value = merging(newHead1,newHead2)
    return value 


def merging (M,L):
    h1 = M 
    h2 = L 
    fH = None 
    fT = None 
    if h1 is None :
        return h2 
    elif h2 is None :
        return h1 
    while h1 and h2 :

        if h1.data < h2.data :
            if fH is None :
                fH = h1 
                fT = h1 
                h1 = h1.next 
            else :
                fT.next = h1 
                fT = fT.next 
                h1 = h1.next
        else :
            if fH is None :
                fH = h2 
                fT = h2 
                h2 = h2.next 
            else :
                fT.next = h2 
                fT = fT.next 
                h2 = h2.next 
    if h1 is not None :
        fT.next = h1 
    else :
        fT.next = h2
    return fH

    





def takeInput():
    inputList = [int(u) for u in input().split()]
    head = None 
    tail = None 
    for curr in inputList:
        if curr == -1 :
            break
        else :
            newNode = Node(curr)
            if head is None:
                head = newNode
                tail = newNode
            else :
                tail.next = newNode
                tail = newNode
    return head 

def printLL(head):
    while head :
        print(head.data,end = ' ')
        head = head.next 
    print()


t = int(input())
while t > 0 :
    head = takeInput()
    value = merge(head)
    printLL(value)
    t-=1
```
