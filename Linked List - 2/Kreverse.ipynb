```python
import sys 
sys.setrecursionlimit(10**6)
class Node :
    def __init__(self,data):
        self.data = data 
        self.next = None 


def takeInput():
    inputList = [int(i) for i in input().split()]
    head = None 
    tail = None 
    for curr in inputList:
        if curr == -1 :
            break
        newNode = Node(curr)
        if head is None :
            head = newNode
        else :
            tail.next = newNode
        tail = newNode
    return head 

def printLL(head):
    while head:  
        print(head.data,end = ' ')
        head = head.next
    print()

def reverse(head):
    if head is None or head.next is None  :
        return head, head
    else :
        smallHead,dbc= reverse(head.next)
        tail = head.next 
        tail.next = head 
        head.next = None
        return smallHead,tail.next

def kReverse(head, k) :
    if head is None or head.next is None :
        return head 
    else :
        h1 = head 
        count = 1 
        t1 = head 
        while count < k :
            if t1.next is None :
                break
            t1 = t1.next 
            count+=1 
        
        h2 = t1.next 
        t1.next = None 
        tempHead, tail = reverse(h1)
        tail.next = kReverse(h2,k)
        return tempHead

t = int(input())
while t> 0 :
    head = takeInput()
    k = int(input())
    value = kReverse(head,k)
    printLL(value)
    t-=1
```
