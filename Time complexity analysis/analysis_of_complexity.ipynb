### triplate sum 

```python
def partition (arr,si,ei):
    pivot = arr[si]
    c = 0 
    for i in range (si,ei+1):
        if arr[i] < pivot :
            c+=1 
    arr[si+c],arr[si] = arr[si],arr[si+c]

    pivot_index = si+c

    i = si 
    j = ei 
    while i < j :
        if arr[i] < pivot :
            i+=1
        elif arr[j] >= pivot :
            j-=1 
        else :
            arr[i],arr[j] = arr[j],arr[i]
            i+=1 
            j-=1
            

    return pivot_index

def quickSort(arr,si,ei) :
    if si >= ei :
        return 
    else :
        pivot_element = partition(arr,si,ei)
        quickSort(arr,si,pivot_element-1)
        quickSort(arr,pivot_element+1,ei)

def triplesum(arr,num,n):
    i = 0 
    count = 0 
    while i < n:
        j = i+1
        k = n 
        while j<k :
            b = arr[j]+arr[k]+arr[i]
            if b < num :
                j+=1
            elif b > num :
                k-=1
            else :
                count +=1
                if arr[k] == arr[k-1] :
                    l = k-1
                    while j < l :
                        b = arr[i]+arr[j]+arr[l]
                        if b == num :
                            count+=1
                            l-=1
                        else :
                            break
                    j+=1
                elif arr[j] == arr[j+1]:
                    j+=1
                    b = arr[i]+arr[j]+arr[k]
                    if b == num :
                        count+=1 
                        j+=1
                    else : 
                        break
                else :
                    j+=1
                    k-=1
        i+=1
    return count

        


t = int(input())
 
while t > 0 :

    n = int(input())
    if n== 0 :
        arr= []
    else :
        arr= [int(k) for k in input().split()]
    num = int(input())
    quickSort(arr,0,n-1)
    print(triplesum(arr,num,n-1))
    t-=1

```

### rotate array 

```python
def rotateArray(arr,n,d):
    temp = []
    for i in range (d):
        temp.append(arr[i])
    i = 0 
    while d < n :
        arr[i] = arr[d]
        i+=1
        d+=1
    
    k = 0 
    while i < n :
        arr[i] = temp[k]
        i+=1
        k+=1
t = int(input())

while t > 0 :
    n = int(input())
    if n==0 :
        arr = []
    else :
        arr= [int(i) for i in input().split()]
    d = int(input())
    rotateArray(arr,n,d)
    for i in arr:
        print(i, end = ' ')
    print()
    t-=1
```

### alternate solution 

```python
def arrayRotate(arr,n,d):

    for i in range (n//2):
        arr[i],arr[n-i] = arr[n-i],arr[i]
    for j in range ((n-d)//2):
        arr[j],arr[n-d-j] = arr[n-d-j],arr[j]
    for k in range (d//2):
        arr[n-d+1+k],arr[n-k] = arr[n-k],arr[n-d+1+k]


n = int(input())
arr= [int(i) for i in input().split()]
d = int(input())
arrayRotate(arr,n-1,d)
print(arr)
```
