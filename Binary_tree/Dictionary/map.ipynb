### hash map class to implement the dictionary 

```python
class MapNode :
    def __init__(self,key,value ):
        self.key = key 
        self.value = value 
        self.next = None 

class Map :
    def __init__(self) :
        self.bucketSize = 5
        self.bucket = [None for i in range(self.bucketSize)]
        self.count = 0 

    def size (self) :
        return self.count
    
    def getVlaue (self, key):
        hc = hash(key)
        index = self.getBucketIndex(hc)
        head = self.bucket[index]
        while head :
            if head.key  == key :
                return head.value 
            head = head.next
        return None 

    def removeValue(self,key):
        hc = hash(key)
        index = self.getBucketIndex(hc)
        head = self.bucket[index]
        prev = None 
        while head :
            if head.key == key :
                if prev is None :
                    self.bucket[index] = head.next 
                else :
                    prev.next = head.next
                self.count-=1
                return head.value
            prev = head
            head = head.next 
        return None  

    def getBucketIndex(self,hc):
        return abs(hc) % (self.bucketSize)

    def rehash(self):
        temp = self.bucket
        self.bucket = [None for i in range(2*self.bucketSize)]
        self.bucketSize  = 2*self.bucketSize
        self.count = 0 
        for head in temp :
            while head :
                self.insert(head.key,head.value)
                head = head.next 
                
    def loadfactor(self):
        return self.count/self.bucketSize

    def insert (self, key , value ):
        hc = hash(key)
        index = self.getBucketIndex(hc)
        head = self.bucket[index]
        while head is not None :
            if head.key == key :
                head.value = value 
                return
            head = head.next 
        head = self.bucket[index]
        newNode = MapNode(key,value)
        newNode.next = head 
        self.bucket[index] = newNode
        self.count+=1
        loadFactor = self.count/self.bucketSize
        if loadFactor > 0.7 :
            self.rehash()    
m = Map()
for i in range(10):
    m.insert("abc"+str(i),i+1)
    print(m.loadfactor())
for i in range(10):
    print("abc"+str(i),m.getVlaue("abc"+str(i)))
# m.insert("aman",10)
# print(m.size())
# print(m.getVlaue("aman"))
# m.insert("khalid",20)
# print(m.size())
# m.insert("aman",90)
# print(m.getVlaue("aman"))
# print(m.size())
# m.insert("junaid",70)
# print(m.size())
# print(m.removeValue("fujail"))
# m.removeValue("aman")
# print(m.size())

```

### pair difference is k 

```python
from math import factorial
def nc2(n):
    ans = 0 
    if n!= 0 and n >= 2 :
        ans = factorial(n)/(factorial(2)*factorial(n-2))
    return int(ans)

def printPairDiffK(l, k):
    count = 0 
    d = {}
    for i in l:
        d[i] = d.get(i,0)+1
    for i in l :
        if k == 0 :
            count = count + nc2(d[i])
            d[i] = 0 
        else :
            temp = i-k 
            temp1 = i+k 
            if temp in d :
                count = count+d[i]*d[temp]
                d[i] = 0 
            if temp1 in d :
                count = count+d[i]*d[temp1]
                d[i] = 0 
    return count

    


n=int(input())
l=list(int(i) for i in input().strip().split(' '))
k=int(input())
print(printPairDiffK(l, k))           
```
